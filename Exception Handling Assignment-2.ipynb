Q1) Explain why we have to inherit the Exception class while creating a custom exception class.

Ans. We inherit exception class to put our custom class in the family of exception classes. This helps us in accessing the methods of exception class that make our custom class compatible with exception handling mechanism i.e. try-except construct.

Q4) Why LookUp error class is used? Explain KeyError and IndexError with an example.

Ans. LookUp class is used to catch excpetions which arises when a key or index is not found in the data structure. KeyError and Index 

```python
my_map={1:10,2:20,3:40}
my_map[4]
# here keyerror occurs in case of dictionary
```

```python
my_list=[1,2,3,4]
my_list[4]
# here IndexError occurs
```

```python
try:
    my_list[4]
except LookupError as e:
    print('An error occured:',e)

# we can also specifically use index error or key error according to our data structure
```

Q6) List down some best practices for exception handling in python.

Ans. Some best practices for exception handling in python are:

    1. Do not use multiple type of except cases. Be specific
    2. Do not use generalized (exception) class for except 
    3. While specifying the cases for except always write the parent error class in the last except
